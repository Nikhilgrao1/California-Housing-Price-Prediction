```python
#Importing the Required Data File
data= pd.read_excel('1553768847_housing.xlsx')
print(data.head())
data.head()
```

# Handle missing values
Fill the missing values with the mean of the respective column.

```python
data.isnull().sum()
```

```python
data['total_bedrooms'].mean()
data['total_bedrooms'].fillna(data['total_bedrooms'].mean(), inplace=True)
data.isnull().sum()
```

Convert categorical column in the dataset to numerical data

```python
#Checking the unique Value in Ocean Proximity
data['ocean_proximity'].value_counts()
data["ocean_proximity"] = data["ocean_proximity"].astype('category')
print(data.dtypes)
data["ocean_proximity"] = data["ocean_proximity"].cat.codes
data.head()
```

# Standardizing  data
Standardize training and test datasets.

```python
from sklearn.preprocessing import StandardScaler
df_standardize = pd.DataFrame(StandardScaler().fit_transform(data),columns=['longitude', 'latitude','housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'ocean_proximity', 'median_house_value'])
df_standardize.head()
```

Extract input (X) and output (Y) data from the dataset.

```python
#Extract input (X) and output (Y) data from the dataset.
x_features = df_standardize[['housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'ocean_proximity']]
y_labels =df_standardize[['median_house_value']]
```

# Split the dataset 
Split the data into 80% training dataset and 20% test dataset.

```python
#Spliting the data into 80% training dataset and 20% test dataset.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train_labels, y_test_labels = train_test_split(x_features,y_labels,train_size=.80,random_state=1)
```

# Performing Linear Regression 
Perform Linear Regression on training data.

```python
from sklearn.linear_model import LinearRegression
logreg = LinearRegression()
logreg.fit(x_train,y_train_labels)
Y_pred = logreg.predict(x_test)
```

Predict output for test dataset using the fitted model

```python
print('actual value = ', y_test_labels.iloc[5])
print('Predicted value is = ', Y_pred[5])
```

Print root mean squared error (RMSE) from Linear Regression.

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_test_labels, Y_pred))
rms
```

# Bonus exercise: Perform Linear Regression with one independent variable

Extract just the median_income column from the independent variables

```python
x_features = df_standardize[['median_income',]]
y_labels =df_standardize[['median_house_value']]
```

Perform Linear Regression to predict housing values based on median_income.

```python
x_train, x_test, y_train_labels, y_test_labels = train_test_split(x_features,y_labels,train_size=.80,random_state=1)
logreg = LinearRegression()
logreg.fit(x_train,y_train_labels)
Y_pred = logreg.predict(x_test)
```

```python
print('actual value = ', y_test_labels.iloc[6])
print('Predicted value is = ', Y_pred[6])
```

```python
rms = sqrt(mean_squared_error(y_test_labels, Y_pred))
rms
```

Plot the fitted model for training data as well as for test data to check if the fitted model satisfies the test data.

```python

plt.scatter(x_train, y_train_labels, color = 'green')
plt.plot (x_train, 
          logreg.predict(x_train), color = 'red')
plt.title ('compare Training result - median_income / median_house_value')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.show()

plt.scatter(x_test, y_test_labels, color = 'blue')
plt.plot (x_test, 
          logreg.predict(x_test), color = 'red')
plt.title ('compare Testing result - median_income / median_house_value')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.show()
```
